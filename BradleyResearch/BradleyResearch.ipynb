```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re

df = pd.read_csv('D:/Research/Distribution/Papers/DistributionPractices/SimFiles/Sequencing/csvs/TallForm100Item.csv')
df["Series ID 1"] = pd.Categorical(df["Series ID 1"],categories = ["PP20, Underloaded","PP20, Balanced","PP20, Overloaded", \
                                                                   "PP26, Underloaded","PP26, Balanced","PP26, Overloaded"])
colors = {10:'black', 100:'black'}

sns.set(rc={'axes.facecolor':'none', 'axes.grid': False, 'axes.linewidth': 1, 'axes.edgecolor':'black', \
            'xtick.color':'black', 'ytick.color': 'black'})

# Initialize a grid of plots
grid = sns.FacetGrid(df, col="Series ID 1", hue = "Number of Items", size = 4, palette = colors, \
                     hue_kws = {"marker": ["o", "s"], "markersize": [7,7]}, col_wrap=3)

# Draw a line plot to show the trajectory of each random walk
grid.map(plt.plot, "Service Level", "% Difference, Random vs. Best Sequence", marker="o", ms=4)

grid.add_legend()
#grid.fig.tight_layout()
plt.subplots_adjust(hspace=0.3)

for ax in grid.axes:
    ax.xaxis.set_tick_params(which = 'both', bottom = 'on', color = 'black', direction = 'out', length = 5, pad = 8, labelsize = 12)
    ax.yaxis.set_tick_params(which = 'both', left = 'on', color = 'black', direction = 'out', length = 5, pad = 8, labelsize = 12)
    ax.xaxis.set_ticklabels
    ax.set_xlabel(ax.get_xlabel(),fontsize = 13)
    ax.set_ylabel(ax.get_ylabel(),fontsize = 13)
    my_title = ax.get_title()
    my_title = re.sub('.*(?=PP)','',my_title)
    ax.set_title(my_title,fontsize = 15)
    if ax.get_ylabel() <> "":
        ax.set_ylabel('% Additional: Random over Best Sequence',fontsize = 12)

grid.savefig("Figure10.jpg", dpi = 1600)  
grid.savefig("Figure10_lowres.jpg", dpi=300) 
```
